```python
import json

with open('db_info.json') as json_file:
    data = json.load(json_file)
```

```python
from pprint import pprint
pprint(data[0])
```

```python
lista_aux = [tweet["fields"] for tweet in data ]
pprint(lista_aux[0])
```

```python
import pandas as pd
df = pd.DataFrame(lista_aux)
df.head()
```

```python
df['fecha']=pd.to_datetime(df['fecha'])#,format='%Y-%m-%dT%H:%M%S')
df.index=df['fecha']
enero = df.groupby(pd.Grouper(freq='M'))
enero.head(20)
```
